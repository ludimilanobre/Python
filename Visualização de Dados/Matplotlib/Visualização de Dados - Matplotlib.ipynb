```python
### TIPOS DE GRÁFICOS ####

# 1) objetivo de transmitir uma única informação importante: único grande 
# número, gráfico de rosca e pictografia.

# 2) objetivo de comparação de valores ou categorias: gráfico de barras,
# gráfico de barras empilhadas, nuvem de palavras, etc.

# 3) objetivo de exibir uma alteração ao longo do tempo ou local: gráfico de 
# linha, gráfico de área, linha do tempo e mapa.

# 4) objetivo de exibir agrupamentos ou rankings de dados: diagramas, 
# mapa mental, carta de fluxo de processos, etc.

# 5) objetivo de exibir relações entre variáveis: dispersão, histograma e 
# multi-série.
```

### Gráfico de Linhas

```python
## [eixo x], [eixo y]
plt.plot([1, 2, 3, 4, 5], [1, 4, 9, 16, 25])
plt.show()
```

```python
## Definindo labels aos eixos:
plt.plot([1, 2, 3, 4, 5], [1, 4, 9, 16, 25])
plt.ylabel("Eixo y")
plt.xlabel("Eixo x")
plt.show()
```

```python
## Importando a biblioteca Numpy:
import numpy as np
```

```python
## Plotando três séries diferentes em um mesmo gráfico:

x1 = np.array(range(1, 11))
y1 = x1

# Série x1 elevada ao quadrado:
x2 = x1
y2 = x1**2

# Série x1 elevada ao cubo:
x3 = x1
y3 = x1**3

## O 'r--' significa que a linha será tracejada e de cor vermelha.
plt.plot(x1, y1, 'r--', x2, y2, 'bs', x3, y3, 'g^')
plt.show()
```

### Dados categorizados

```python
## Gráfico de colunas:

grupos = ['Produto A', 'Produto B', 'Produto C']
valores = [1, 10, 100]

plt.bar(grupos, valores)
plt.show()
```

```python
## Gráficos de colunas juntas:

valores_A = [6, 7, 8, 4, 4]
valores_B = [3, 12, 3, 4, 6]

# Cria eixo X para o produto A e produto B com uma separação de 0.25 entre as
# barras.
x1 = np.arange(len(valores_A))
x2 = [x + 0.25 for x in x1]

## Plotando as barras:
plt.bar(x1, valores_A, width = 0.25, label = "Produto A", color = 'b')
plt.bar(x2, valores_B, width = 0.25, label = "Produto B", color = 'y')

meses = ['Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']
plt.xticks([x + 0.12 for x in range(len(valores_A))], meses)

plt.legend()

plt.title("Vendas")
plt.show()
```

### Gráfico de Pizza

```python
vendas = [3000, 2300, 1000, 500]
labels = ['E-commerce', 'Loja Física', 'E-mail', 'Marketplace']

plt.pie(vendas, labels = labels)
plt.show()
```

```python
## Nível de separabilidade entre as partes.
## A ordem do vetor representa os labels.
explode = (0.1, 0, 0, 0)

plt.pie(vendas, labels = labels, explode = explode)
plt.show()
```

```python
## Cálculo dos percentuais para plotar no gráfico:
plt.pie(vendas, labels = labels, explode = explode, autopct = '%1.1f%%')
plt.show()
```

```python
## Legenda:
## o parâmetro 'loc' indica a posição da legenda. Pode ser igual a 1, 2, 3, 4.
## o parâmetro 'shadow' dá um aspecto 3D para o gráfico.
plt.pie(vendas, explode = explode, autopct = '%1.1f%%', shadow = True)
plt.legend(labels, loc = 2)
## plt.legend(labels, loc = 'best')

## Para o gráfico não ficar desproporcional:
plt.axis('equal')

plt.show()

## Melhor gráfico de pizza EVER!
```

### Matplotlib, pyplot e pylab?

```python
### PYPLOT: é um módulo do pacote Matplotlib que implicitamente cria figuras e
### eixos nos gráficos.


### PYLAB: é outro módulo do Matplotlib que já importa o Numpy e o Pyplot. Este
### módulo é projetado para plots de arrays e cálculos matemáticos. 
### Porém está depreciado.
```

```python
import matplotlib.pyplot as plt
import numpy as np

## Criando uma figura:
fig = plt.figure()

## Definindo os eixos:
ax = fig.add_subplot(111)

## Scatter plot usando o Numpy:
ax.scatter(np.linspace(0, 1, 5), np.linspace(0, 5, 5))

plt.show()
```

```python
### SUBPLOTS: deve-se definir o número de linhas e o número de colunas.
## As colunas indicam a quantidade de gráficos em cada linha.

plt.subplots(nrows = 1, ncols = 2)
```

```python
## Para acessar subplots utiliza-se o método 'subplot':
x = [1, 2, 3, 4]
y = [10, 20, 25, 40]

plt.subplot(1, 2, 1)
plt.plot(x, y, 'o')

plt.subplot(1, 2, 2)
plt.plot(x, y, color = 'r')
```

```python
### Criação de instâncias da classe figure:

## Cria um objeto do tipo figure:
fig = plt.figure()

## Define o tamanho em percentual do gráfico.
## Altura e largura do eixo na figura.
axes_a = fig.add_axes([0.1, 0.1, 0.8, 0.8])

## Para plotar o gráfico:
axes_a.plot(x, y)
```

```python
### Figura com dois gráficos de proporções diferentes.

fig = plt.figure()
axes_a = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes_b = fig.add_axes([0.2, 0.5, 0.3, 0.3])
axes_a.plot(x, y)
axes_b.plot(x, y, 'r--')
```

```python
### Customização da altura e largura do gráfico.

import matplotlib.pyplot as plt
import numpy as np

x = [1, 2, 3, 4]
y = [10, 20, 30, 40]

## Altera a largura e altura do gráfico (width, height).
fig, ax = plt.subplots(figsize = (20, 5))

## Para plotar o gráfico:
ax.plot(x, y, 'r--')
```

```python
### Dois gráficos (eixos) na mesma figura.

## Definindo as configurações dos plots: cada plot terá o mesmo tamanho de
## figuras.
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (10, 5))

## Dados para cada subplot:
ax1.bar([1, 2, 3], [3, 4, 5])
ax2.barh([0.5, 1, 2.5], [0, 1, 2])

plt.show()
```

```python
### Configurações dos eixos.

fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (10, 5))

## Dados para cada subplot:
ax1.bar([1, 2, 3], [3, 4, 5], color = 'r')
ax2.barh([0.5, 1, 2.5], [0, 1, 2])

## Método 'set' define as configurações do gráfico.
ax1.set(title = 'Gráfico de barras verticais', xlabel = 'Eixo x', ylabel = 'Eixo y')
ax2.set(title = 'Gráfico de barras horizontais', xlabel = 'Eixo x', ylabel = 'Eixo y')

plt.show()
```

### Adicionar textos em gráficos

```python
## Gerar 100 valores no intervalo de 0 a 2:
x = np.linspace(0, 2, num = 100)

plt.plot(x, x, label = 'Linear')
plt.plot(x, x**2, label = 'Quadrático', linewidth = 5)

plt.xlabel('X')
plt.ylabel('Y')

plt.title('Gráfico de linhas simples')


## Exibe a legenda e por padrão usa o label de cada plot.
plt.legend()


## Configurações do texto:
plt.text(1, 1, 'Cruzamento das linhas', fontsize = 10, horizontalalignment = 'right')

plt.show()
```

### Dataframes Pandas

```python
## Importação do Pandas.
import pandas as pd
```

```python
## Dicionário de valores:
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.random(50)
       }

data = pd.DataFrame(data)
data.head()
```

```python
## Gerando dados aleatórios para 'b' e 'd':
data['b'] = data['a'] + 10*np.random.random(50)
data['d'] = np.abs(data['d'])*200
```

```python
## Gráfico de dispersão com o pandas dataframe. 
## Parâmetro 'c' = cor do ponto.
## Parâmetro 's' = tamanho do ponto.
plt.scatter('a', 'b', c = 'c', s = 'd', data = data)
plt.xlabel('A')
plt.ylabel('B')
plt.show()
```
