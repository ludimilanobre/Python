## Seaborn

```python
## Interface de alto nível para gráficos estatísticos.
## Simples e intuitiva.
## Galeria do Seaborn: http://seaborn.pydata.org/examples/index.html.

## Quando utilizar? Análise exploratória de dados e apresentação de resultados.
```

```python
## Importação:
import seaborn as sns
```

```python
df_tips = sns.load_dataset('tips')
df_tips.head()
```

```python
### Função 'relplot()': método que combina o scatter plot e o line plot.
## É útil para ver relações entre duas ou mais variáveis com o uso de
## parâmetros de semântica dos dados.

tips = df_tips
sns.relplot(x = 'total_bill', y = 'tip', data = tips);
```

```python
## Visualizando a relação entre as variáveis 'total_bill' e 'tip' ao colorir
## cada ponto com a variável 'smoker' a partir do parâmetro 'hue'.
sns.relplot(x = 'total_bill', y = 'tip', hue = 'smoker', data = tips);
```

```python
## Visualizando a relação entre as variáveis 'total_bill' e 'tip' ao colorir
## cada ponto com a variável 'sex' a partir do parâmetro 'hue'.
sns.relplot(x = 'total_bill', y = 'tip', hue = 'sex', data = tips);
```

```python
## Visualizando a relação entre as variáveis 'total_bill' e 'tip' ao colorir
## cada ponto com a variável 'sex' a partir do parâmetro 'hue'. O estilo também
## será definido pela variável 'sex'.
sns.relplot(x = 'total_bill', y = 'tip', hue = 'sex', style = 'sex', data = tips);
```

```python
## Representação da relação de quatro variáveis.
sns.relplot(x = 'total_bill', y = 'tip', hue = 'sex', style = 'smoker', 
           data = tips);
```

```python
## Paleta de cores segundo variável categórica ('size').
sns.relplot(x = 'total_bill', y = 'tip', hue = 'size', data = tips);
```

```python
## Paleta de cores seaborn: 
## https://seaborn.pydata.org/generated/seaborn.color_palette.html
```

```python
## Parâmetro 'size' é o tamanho de cada ponto.
sns.relplot(x = 'total_bill', y = 'tip', size = 'size', data = tips);
```

```python
sns.relplot(x = 'total_bill', y = 'tip', size = 'size', sizes = (15, 300),
           data = tips);
```

### Cores no Seaborn

```python
## A função que controla as cores no Seaborn é a 'color_palette()'.
## Estilo padrão: 'set_palette()'.

## Ferramenta Color Brewer 2.0:
## https://colorbrewer2.org/#type=sequential&scheme=BuGn&n=3
```

```python
## Paleta de cores para dados quantitativos:
sns.palplot(sns.color_palette('Accent'))
```

```python
## Paleta de cores para dados divergentes:
sns.palplot(sns.color_palette('PuOr'))
```

```python
## Paleta de cores para dados sequenciais:
sns.palplot(sns.color_palette('Blues'))
```

```python
### Aplicando a paleta de cor 'Blues' a um gráfico:
sns.relplot(x = 'total_bill', y = 'tip', hue = 'size', palette = 'Blues',
           size = 'size', sizes = (12, 250), data = tips);
```

```python
### Gráfico com paleta de cor qualitativa.
sns.catplot(x = 'sex', kind = 'count', palette = 'Set2', data = tips);
```

```python
### Gráfico com paleta de cor divergente.
sns.heatmap(tips.corr(), annot = True, cmap = 'PiYG');
```

```python
## Paleta de cores padrão para gráficos do Seaborn.
sns.set_palette('pastel')

sns.relplot(x = 'total_bill', y = 'tip', data = tips);
```

```python
### Instrução 'with' para definir cor temporariamente.

with sns.color_palette('Accent'):
    sns.relplot(x = 'total_bill', y = 'tip', size = 'size', sizes = (12, 250),
               data = tips);
```

### Dados categóricos

```python
## O Seaborn fornece o método 'catplot()' que já contém diversos tipos de 
## gráficos embutidos.
```

```python
### Gráfico de barras ou colunas:
sns.catplot(x = 'sex', kind = 'count', palette = 'Set2', data = tips);
```

```python
### Gráfico de barras horizontais:
sns.catplot(y = 'day', kind = 'count', palette = 'Set1', data = tips);
```

```python
### Scatter plot com dados categóricos:
sns.catplot(x = 'day', y = 'total_bill', palette = 'Set2', data = tips);
```

```python
## Parâmetro 'swarm' evita sobreposição de pontos.
sns.catplot(x = 'day', y = 'total_bill', kind = 'swarm', data = tips);
```

```python
## Parâmetro 'hue' permite adicionar uma terceira variável na análise.
sns.catplot(x = 'day', y = 'total_bill', kind = 'swarm', hue = 'sex',
           palette = 'Dark2', data = tips);
```

```python
## O parâmetro 'order' permite alterar a ordem padrão das categorias que estão
## sendo exibidas. Essa opção é útil quando há mais de um gráfico na figura e
## deseja-se manter o mesmo ordenamento em ambos.
sns.catplot(x = 'day', y = 'total_bill', kind = 'swarm', hue = 'sex',
           order = ['Sat', 'Sun', 'Fri', 'Thur'], data = tips);
```

```python
## Invertendo a visualização para horizontal.
sns.catplot(x = 'total_bill', y = 'day', kind = 'swarm', data = tips);
```

### Gráficos com regressão

```python
## Relação entre variáveis quantitativas através de regressão.
## Com modelos de regressão simples é possível verificar se existe alguma
## correlação entre variáveis.
```

```python
## Linha de regressão para visualizar correlação entre as variáveis.
sns.lmplot(x = 'total_bill', y = 'tip', hue = 'sex', palette = 'Pastel1',
          data = tips);
```

```python
## Parâmetro 'col' para segregar os gráficos pelo valor da coluna 'time'.
sns.lmplot(x = 'total_bill', y = 'tip', hue = 'smoker', col = 'time',
          palette = 'Set2', data = tips);
```

```python
## Parâmetros 'scatter_kws' e 'line_kws'.
## 'scatter_kws': 's' é o tamanho do ponto; 'alpha' é o nível de opacidade da cor.
## 'line_kws': 'lw' é a largura da linha.

sns.lmplot(x = 'total_bill',
           y = 'tip',
         hue = 'smoker',
         col = 'time',
     palette = 'Set1',
        data = tips,
 scatter_kws = {'s': 90, 'alpha': 0.5},
    line_kws = {'lw': 4});
```

```python
## Segregação pelos parâmetros 'row' e 'col'.
sns.lmplot(x = 'total_bill', y = 'tip', hue = 'smoker', col = 'time',
          row = 'sex', palette = 'Set2', data = tips);
```

### Gráficos de distribuição

```python
### BOXPLOT:
sns.catplot(x = 'day', y = 'total_bill', kind = 'box', data = tips);
```

```python
sns.catplot(x = 'day', y = 'total_bill', hue = 'smoker', kind = 'box', 
           data = tips);
```

```python
### BOXER: tipo de boxplot com foco maior na forma dos dados e não nos outliers.
sns.catplot(x = 'day', y = 'total_bill', kind = 'boxen', data = tips);
```

```python
## O método 'stripplot' permite plotar a distribuição dos dados.
## boxen + stripplot podem ser combinados para ter mais informação.
sns.catplot(x = 'day', y = 'total_bill', kind = 'boxen', data = tips);
sns.stripplot(x = 'day', y = 'total_bill', data = tips, color = 'gray');
```

```python
sns.catplot(x = 'sex', y = 'total_bill', hue = 'smoker', col = 'time',
           data = tips, kind = 'boxen', height = 4);
```

```python
### VIOLIN PLOT:
sns.catplot(x = 'day', y = 'total_bill', kind = 'violin', data = tips);
```

```python
sns.catplot(x = 'total_bill', y = 'day', kind = 'violin', data = tips);
```

```python
## Parâmetro 'hue' para segregar os dados por uma terceira variável.
sns.catplot(x = 'day', y = 'total_bill', hue = 'time', kind = 'violin',
           data = tips);
```

```python
## Parâmetro 'split' para dividir o gráfico segundo uma variável.
sns.catplot(x = 'day', y = 'total_bill', hue = 'sex', split = True, 
           kind = 'violin', data = tips);
```

```python
## Parâmetro 'inner' para preencher o violino com observações.
## É útil para destacar as partes que possuem mais dados.
sns.catplot(x = 'day', y = 'total_bill', hue = 'sex', split = True, 
           inner = 'stick', kind = 'violin', data = tips);
```

```python
## Combinação do violin plot com 'swarmplot'.
sns.catplot(x = 'day', y = 'total_bill', hue = 'sex', split = True,
           kind = 'violin', data = tips);
sns.swarmplot(x = 'day', y = 'total_bill', color = 'k', size = 3, data = tips);
```

### Relacionamentos emparelhados

```python
## Plot de múltiplas variáveis de um dataset.
## Cria uma matriz de eixos para mostrar a relação de cada par de colunas no
## dataframe.
## Por padrão também plota um histograma de cada coluna do dataframe na 
## diagonal da matriz.
```

```python
## Somente para dados numéricos.
sns.pairplot(tips);
```

```python
## Dataset 'iris':
iris = sns.load_dataset('iris')
iris.head()
```

```python
## Pares emparelhados segregando pela variável 'species':
sns.pairplot(iris, hue = 'species', palette = 'Set1');
```

```python
## Histogramas na diagonal:
sns.pairplot(iris, hue = 'species', diag_kind = 'hist');
```

```python
## Plotando apenas duas colunas do dataframe:
sns.pairplot(iris, hue =  'species', vars = ['sepal_width', 'sepal_length'],
            palette = 'husl');
```

```python
## Scatter plot com regressão:
sns.pairplot(iris, kind = 'reg');
```
